How are in-demand skills trending for Data Analysts?
Methology
Aggregate skill counts monthly
Re-analyze based on percentage of total jobs
Plot the monthly skill demand

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US_DA = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_US_DA['job_posted_month_no'] = df_US_DA['job_posted_date'].dt.month
```

```python
df_US_DA_explode = df_US_DA.explode('job_skills')
```

```python
df_US_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_US_DA_pivot = df_US_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_US_DA_pivot.loc['Total'] = df_US_DA_pivot.sum()
```

```python
df_US_DA_pivot = df_US_DA_pivot[df_US_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_US_DA_pivot = df_US_DA_pivot.drop('Total')
```

```python
df_US_DA_pivot
```

```python
df_US_DA.groupby('job_posted_month_no').size()
```

```python
DA_totals = df_US_DA.groupby('job_posted_month_no').size()
```

```python
# divide first 12 rows of df_DA_pivot by DA_totals
df_DA_US_percent = df_US_DA_pivot.div(DA_totals/100, axis= 0)

# changes month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]
```

```python
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

#add name to the lines
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
