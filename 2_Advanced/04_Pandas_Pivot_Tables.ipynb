```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Count of Job Titles

```python
df.pivot_table(index='job_title_short',aggfunc='size')
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
df.pivot_table(values='salary_year_avg', index='job_country' , columns='job_title_short' , aggfunc='median')
```

```python
df_job_country_salary = df.pivot_table(
    values='salary_year_avg',        # The data to aggregate
    index='job_country',             # Rows (index of pivot table)
    columns='job_title_short',       # Columns (pivot table headers)
    aggfunc='median'                 # Aggregation function
)

df_job_country_salary
```

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
#this filterd the data to the top 6 countries
```

```python
df_job_country_salary = df_job_country_salary.loc[top_countries]
df_job_country_salary
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

#no need to use loc or iloc since we are filtering on header
df_job_country_salary = df_job_country_salary[job_titles]
df_job_country_salary
```

```python
df_job_country_salary.plot(kind='bar')
```

```python
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')  
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
ax = df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')

# Extract United States values for each job title
us_salaries = df_job_country_salary.loc['United States']

# Loop through job titles and plot a horizontal line for each
colors = ['C0', 'C1', 'C2']  # Default matplotlib colors for first three bars
for idx, (job_title, salary) in enumerate(us_salaries.items()):
    plt.axhline(
        y=salary,
        color=colors[idx], 
        linestyle='dotted',
        linewidth=2,
        label=f'US {job_title} Median'
    )

plt.show()
```
